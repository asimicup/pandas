# Accessing the a group of rows and columns by labels or a boolean array

we perform data selecting operatoins throug loc and iloc

loc => accessing rows and columns using labels
iloc => accessing rows and columns using their numbers (integer location based indexing)

```python
import pandas as pd
```

```python
df = pd.read_csv('D:\\Projects\\appledata.csv')
df
```

```python
df.loc[0]
```

```python
df.loc[[0,1]]
```

```python
df.loc[4, 'Length']
```

```python
df.loc[0:3, 'Pixel count']
```

```python
df.loc[df['Fruit'] < 1]
```

```python
df.iloc[0]
```

```python
df.iloc[[0]]
```

```python
df.iloc[:, 0]
```

```python
df.iloc[[0,1]]
```
