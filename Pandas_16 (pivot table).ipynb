#### The levels in the pivot table will be stored in multi-indexed objects (hierarical indexes) on the indexe and columns of the result DataFrame

```python
import pandas as pd
```

```python
df = pd.read_csv('D:\Projects\headbrain.csv')
df
```

```python
df.pivot_table(index = 'Gender')
```

```python
df.pivot_table(index = 'Gender', columns = 'Age Range')
```

```python
df.pivot_table(index = 'Gender', columns = 'Age Range', aggfunc = 'sum')
```
