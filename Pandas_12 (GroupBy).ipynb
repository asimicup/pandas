### Splitting the data into groups based on some criteria

a groupby operation involves one of the following operations on the original object

1. splitting the object
2. applying a function
3. combining the results

```python
import pandas as pd
```

```python
df = pd.read_csv('D:\\Projects\\car_price.csv')
df
```

```python
grp1 = df.groupby(by = 'body_type')
grp1
```

```python
grp1.groups
```

```python
grp1 = df.groupby(by = ['body_type', 'hp']).groups
grp1
```

```python
for body_type, hp in grp1:
    print(body_type)
    print(hp)
```

```python
list(grp1)
```

```python
dict(list(grp1))
```

```python
grp2 = df.groupby('body_type').get_group('Station wagon')
grp2
```

```python
grp2.describe()
```

```python
grp2.sum()
```

```python
grp2.mean()
```

```python
grp2.agg(['sum', 'max', 'mean'])
```
